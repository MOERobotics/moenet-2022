```python
import pandas as pd
df = pd.read_csv('testing62.csv', index_col=0)
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots()
row3 = df.loc[3]
row3 = row3[row3['pose_Z'] < 80]
row3 = row3[50:]
print(np.std(row3))
avg = np.average(row3['pose_Z'])
ax.plot([i for i in range(len(row3))],row3['pose_Z'], label='ID: 3')
ax.plot([i for i in range(len(row3))], [avg for i in range(len(row3))], label='Avg')
ax.set_ylabel('Distance (in.)')
ax.set_xlabel('Frame')
ax.set_title('Variation in Distance Over Frames @ ~21 ft.')
```

```python
plt.hist(row3['pose_Z'],100)
plt.axvline(avg,color='orange', linestyle='--')
plt.xlabel('Distance (in.)')
plt.ylabel('Count')
plt.title('Distribution of Measured Distance @ ~21 ft.')
plt.show()
```
